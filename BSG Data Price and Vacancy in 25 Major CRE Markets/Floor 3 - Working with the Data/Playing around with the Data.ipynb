```python
import pandas as pd

# Load the CSV file
df = pd.read_csv('extracted_data.csv')

# Function to format listing rate as dollars and cents
def format_listing_rate(val):
    return f"${val:.2f}"

# Function to format vacancy rate as a percentage
def format_vacancy_rate(val):
    return f"{val:.1f}%"

# Apply formatting
df['Listing Rate'] = df['Listing Rate'].apply(format_listing_rate)
df['Total Vacancy Rate'] = df['Total Vacancy Rate'].apply(format_vacancy_rate)

# Print the updated DataFrame
print(df)

# Save the updated DataFrame to a new CSV file
df.to_csv('formatted_data.csv', index=False)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.dates import YearLocator, DateFormatter

# Read the updated CSV file
df = pd.read_csv('formatted_data.csv')

# Create a 'Date' column from 'Year' and 'Month'
df['Date'] = pd.to_datetime(df[['Year', 'Month']].assign(DAY=1))

# Convert 'Total Vacancy Rate' to numeric (removing % sign)
df['Total Vacancy Rate'] = df['Total Vacancy Rate'].str.rstrip('%').astype('float') / 100

# Plot Vacancy Rate over time for Bay Area
bay_area_vacancy_data = df[df['Market'] == 'Bay Area'].sort_values('Date')

plt.figure(figsize=(12, 6))
plt.plot(bay_area_vacancy_data['Date'], bay_area_vacancy_data['Total Vacancy Rate'], '-o')

plt.title('Bay Area Vacancy Rate Over Time')
plt.xlabel('Date')
plt.ylabel('Vacancy Rate')

# Format x-axis to show only years
plt.gca().xaxis.set_major_locator(YearLocator())
plt.gca().xaxis.set_major_formatter(DateFormatter('%Y'))

# Format y-axis as percentage
plt.gca().yaxis.set_major_formatter(plt.FuncFormatter(lambda y, _: f'{y:.0%}'))

plt.grid(True, linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()

# Plot Listing Rate over time for Bay Area
# Convert 'Listing Rate' to numeric (removing $ sign)
df['Listing Rate'] = df['Listing Rate'].str.replace('$', '').astype('float')

plt.figure(figsize=(12, 6))
plt.plot(bay_area_vacancy_data['Date'], bay_area_vacancy_data['Listing Rate'], '-o')

plt.title('Bay Area Listing Rate Over Time')
plt.xlabel('Date')
plt.ylabel('Listing Rate ($)')

# Format x-axis to show only years
plt.gca().xaxis.set_major_locator(YearLocator())
plt.gca().xaxis.set_major_formatter(DateFormatter('%Y'))

plt.grid(True, linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
